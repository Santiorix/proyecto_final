```python
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
ltyss = pd.read_csv('../output/ltyss.csv')
consumos = pd.read_csv('../output/consumos_actuales.csv')
```

```python
ltyss.head(1000)
```

```python
consumos.head(1000)
```

```python
#voy a hacer un unico df con el consumo, el lt y el ss para calcular el punto de pedido
consolidado = consumos.merge(ltyss, on='id_mmpp', how='left')
```

```python
consolidado.head(1000)
```

```python
consolidado.info() # me tengo que quitar los nulos, por que si no tienen ni stock de seguridad ni plazo de entrega es que no van por MRP
```

```python
consolidado = consolidado.dropna()
```

```python
consolidado.head(1000) #Ya tengo las mmpp que quiero con los valores
```

```python
#ahora voy a calcular el punto de pedido
consolidado["punto_pedido"] = consolidado["Stock de seguridad"] + consolidado["Consumo_diario"]*consolidado["Plazo de entrega"]
```
