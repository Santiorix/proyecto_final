```python
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
maestro_mat = pd.read_excel('../../BBDD_proyecto_final/Maestro_Materiales.xlsx', sheet_name = "Maestro completo", index_col = False)
```

```python
maestro_mat.head()
```

```python
#convierto el centro en string para poder concatenar
maestro_mat["Centro"] = maestro_mat["Centro"].apply(str)
```

```python
#concateno centro y material para sacar el mismo id que en el otro notebook
maestro_mat["id_mmpp"] = maestro_mat["Centro"] + "_" + maestro_mat["Material"]
```

```python
maestro_mat.head()
```

```python
# me hago una pt para quedarme solo con id, stock de seguridad y plazo de entrega
ltyss = pd.pivot_table(maestro_mat, values=['Stock de seguridad', "Plazo de entrega"], index=['id_mmpp'], aggfunc=np.mean)
```

```python
ltyss.head(1000)
```

```python
ltyss.to_csv("../output/ltyss.csv")
```
