```python
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
stocks_2017 = DataFrame()#Uno todos las hojas del libro stocks 2017 en un solo DF
for i in range (12):
    df_stocks = pd.read_excel('../BBDD_proyecto_final/Stocks_2017.xlsx', sheet_name = i, index_col = False)
    stocks_2017 = stocks_2017.append(df_stocks)
    
```

```python
stocks_2017.reset_index(inplace = True)
#stocks_2017
```

```python
stocks_2018 = DataFrame()
for i in range (12):
    df_stocks = pd.read_excel('../BBDD_proyecto_final/Stocks_2018.xlsx', sheet_name = i, index_col = False)
    stocks_2018 = stocks_2018.append(df_stocks)
```

```python
stocks_2018.reset_index(inplace = True)
```

```python
#stocks_2018
```

```python
stocks_2019 = DataFrame()
for i in range (8):
    df_stocks = pd.read_excel('../BBDD_proyecto_final/Stocks_2019.xlsx', sheet_name = i, index_col = False)
    stocks_2019 = stocks_2019.append(df_stocks)
```

```python
stocks_2019.reset_index(inplace = True)
```

```python
#stocks_2019
```

```python
Stocks_1 = stocks_2017.append(stocks_2018)
```

```python
#Stocks_1
```

```python
Stocks_total = Stocks_1.append(stocks_2019)
```

```python
Stocks_total.reset_index(inplace = True)
```

```python
Stocks_total["Total cantidades salida"]=Stocks_total["Total cantidades salida"].str.replace("-","")#limpio la columna cantidades de salida
```

```python
Stocks_total["Total cantidades salida"]=Stocks_total["Total cantidades salida"].fillna(0)#Quito los NA sustituyendolos por cero
```

```python
Stocks_total["Valor inicial"]=Stocks_total["Valor inicial"].str.replace("-","") #repito el proceso para las columnas con guion
Stocks_total["Valor inicial"]=Stocks_total["Valor inicial"].fillna(0)
Stocks_total["Valor final"]=Stocks_total["Valor final"].str.replace("-","")
Stocks_total["Valor final"]=Stocks_total["Valor final"].fillna(0)
```

```python
Stocks_total= Stocks_total.drop(["level_0", "index"], axis = 1) #Limpio las columnas que no me sirvenº
```

```python
Stocks_total["De fecha"] = Stocks_total["De fecha"].str.replace(".","/")
Stocks_total["A fecha"] = Stocks_total["A fecha"].str.replace(".","/")
```

```python
Stocks_total["De fecha"] = pd.to_datetime(Stocks_total["De fecha"], dayfirst=True)#convierto las columnas fecha de string a fecha
Stocks_total["A fecha"] = pd.to_datetime(Stocks_total["A fecha"], dayfirst=True)
```

```python
Stocks_total["días_mes"] = Stocks_total["A fecha"]-Stocks_total["De fecha"]
```
